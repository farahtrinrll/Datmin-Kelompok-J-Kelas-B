```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas import read_csv
dfmerge
dfmerge.shape
dfmerge.info()
```

```python
dfmerge = pd.concat([dfJ,dfB,dfD,dfT,dfX])
dfmerge
```

```python
dfmerge.drop(dfmerge.columns[14:26], axis=1, inplace=True)
print(dfmerge)
```

```python
dfmerge.duplicated().sum()
```

```python
#melihat adanya outlier
p = sns.boxplot(x="LT", y="harga", data=dfmerge)
#mengecek apakah terdapat duplikasi data
print(dfmerge.shape)
dfmerge.duplicated().sum()
```

```python
#menghapus data duplikat
dfmerge.drop_duplicates(inplace=True)
print(dfmerge.duplicated().sum())
print(dfmerge.shape)
```

```python
#untuk mengetahui harga terbesar terdapat di lokasi mana
besar_harga=dfmerge.groupby('lokasi')['harga'].sum().reset_index()
besar_harga
```
